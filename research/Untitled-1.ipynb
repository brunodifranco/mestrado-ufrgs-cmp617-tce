```python
import os
import json
import pandas as pd
os.chdir("..")
```

```python
path = 'src/lda_opt_outputs'

data = []

for file in os.listdir(path):
    if file.endswith('.json'):
        full_path = os.path.join(path, file)
        with open(full_path, 'r') as f:
            json_data = json.load(f)

            item = {
                "nlp_normalization_method": json_data["nlp_normalization_method"],
                "top_n": json_data["topn"],
                "n_filter": json_data["params"]["n_filter"],
                "num_topics": json_data["params"]["num_topics"],
                "chunksize": json_data["params"]["chunksize"],
                "passes": json_data["params"]["passes"],
                "alpha": json_data["params"]["alpha"],
                "eta": json_data["params"]["eta"],
                "decay": json_data["params"]["decay"]
            }
            data.append(item)
```

```python
df = pd.DataFrame(data).sort_values(by=["nlp_normalization_method", "top_n"])
```
